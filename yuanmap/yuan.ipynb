```python
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Map
```

```python
# 您的数据（直接从Excel复制，西藏无数据）
data = [
    ("北京", 7), ("天津", 5), ("河北", 6), ("山西", 4), ("内蒙古", 3),
    ("辽宁", 6), ("吉林", 4), ("黑龙江", 5), ("上海", 2), ("江苏", 2),
    ("浙江", 4), ("安徽", 5), ("福建", 3), ("江西", 4), ("山东", 6),
    ("河南", 5), ("湖北", 4), ("湖南", 5), ("广东", 4), ("广西", 3),
    ("海南", 3), ("重庆", 4), ("四川", 8), ("贵州", 2), ("云南", 2),
    ("陕西", 3), ("甘肃", 2), ("青海", 2), ("宁夏", 2), ("新疆", 4)
]

# 创建地图
c = (
    Map()
    .add(
        series_name="数值",
        data_pair=data,
        maptype="china",
        is_map_symbol_show=False,  # 隐藏标记点
        label_opts=opts.LabelOpts(is_show=True, formatter="{b}: {c}"),  # 显示省份和数值
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="中国各省数值分布"),
        visualmap_opts=opts.VisualMapOpts(
            min_=2,  # 最小值
            max_=8,  # 最大值
            range_color=["#E6F2FF", "#0059b3"],  # 蓝色渐变
            is_piecewise=False,  # 连续渐变
        ),
    )
)

# 生成HTML文件（用浏览器打开）
c.render("china_map.html")
```
